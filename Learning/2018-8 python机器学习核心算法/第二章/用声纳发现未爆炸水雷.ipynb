## 确定数据集规模

```python
import urllib2
import sys
```

```python
target_url = ('http://archive.ics.uci.edu/ml/machine-learning-'
              'databases/undocumented/connectionist-bench/sonar/sonar.all-data')
data = urllib2.urlopen(target_url)
```

```python
xList = []
labels = []
for line in data:
    row = line.strip().split(',')
    xList.append(row)
sys.stdout.write('Num of rows: '+str(len(xList))+'\n')
sys.stdout.write('Num of Columns: '+str(len(xList[1])))
```

## 确定每个特征的属性

```python
import pandas as pd
nrow = len(xList)
ncol = len(xList[0])
```

```python
floatcol = []
strcol = []
for col in range(ncol):
    try:
        if isinstance(float(xList[col][0]),float):
            floatcol.append(str(col))
    except:
        strcol.append(str(col))
sys.stdout.write('The cols of float are: '+','.join(floatcol)+'\n')
sys.stdout.write('The cols of str are: '+','.join(strcol)+'\n')
```

## 用分位图展示异常点

```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
stats.probplot([float(x) for x in list(xList[3])], dist='norm', plot = pylab)
pylab.show()
```

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-5, 5, 0.1)
y = stats.norm.cdf(x, 0, 1)
plt.plot(x, y)
plt.show()
```
